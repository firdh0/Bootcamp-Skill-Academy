# üéØ **Step 1: Load model**
---

```python
model = YOLO("yolo11n-cls.pt")
```

```python
path ='/content/drive/MyDrive/SKILL ACADEMY/PROJECT 8/Dataset_v2'
save_dir = '/content/drive/MyDrive/SKILL ACADEMY/PROJECT 8/Dataset_v2'
```

# üéØ **Step 2: Train model**

```python
results = model.train(data=path, epochs=10, project=save_dir, name="YOLO_training")
```
