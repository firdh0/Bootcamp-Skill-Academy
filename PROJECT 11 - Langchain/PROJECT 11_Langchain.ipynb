```python
import openai
from google.colab import userdata

OPENAI_API_KEY = userdata.get("OPENAI_API_KEY")

OPENAI_API_KEY
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.document_loaders import TextLoader
from langchain.vectorstores import FAISS
```

```python
from langchain.document_loaders import TextLoader

loader = TextLoader("/content/The Project Gutenberg.txt")
documents = loader.load()
text_splitter = RecursiveCharacterTextSplitter(chunk_size=700, chunk_overlap=50)
docs = text_splitter.split_documents(documents)

embeddings = OpenAIEmbeddings(openai_api_key=OPENAI_API_KEY)
```

```python
db = FAISS.from_documents(docs, embeddings)
```

```python
from langchain.chains import RetrievalQA
from langchain.chat_models import ChatOpenAI
```

```python
OpenAIModel = "gpt-3.5-turbo"
llm = ChatOpenAI(model=OpenAIModel, temperature=0.1, openai_api_key=OPENAI_API_KEY)

qa = RetrievalQA.from_chain_type(llm=llm, retriever=db.as_retriever())
```

```python
query = "please mention all of the characters?"

qa.run(query)
```

```python
query = "What are the full names of the two main characters in Pride and Prejudice?"

_answer = qa.run(query)
_answer
```
